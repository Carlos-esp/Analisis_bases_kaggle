#### Limpiamos corchetes

```python
x=nflx['Premiere'][0]
y=nflx['Premiere'][117]
print(y)
print(re.findall('.*(?=\[)|.*',y))
print(x)
print(re.findall('.*(?=\[)|.*',x))
```

```python
def eliminar_corch(x):
    if pd.isnull(x):
        return np.nan
    else:
        return re.findall('.*(?=\[)|.*',x)[0]
```

```python
# nflx['Title']=nflx['Title'].apply(eliminar_corch)
for d in nflx.columns:
    nflx[d]=nflx[d].apply(eliminar_corch)
```

```python
nflx[nflx['Genre'].isna()]
```

#### Obtenemos nuevas columnas con las fechas

```python
x=nflx['Premiere'][0]
print(re.findall('\w*(?=\s)',x)) #mes
print(re.findall('(?<=\s)\d*(?=,)',x)) #dia
print(re.findall('\d*$',x)) #año
```

```python
def mes(x):
    if pd.isnull(x):
        return np.nan
    else:
        return re.findall('\w*(?=\s)',x)[0]

def dia(x):
    if pd.isnull(x):
        return np.nan
    else:
        return re.findall('(?<=\s)\d*(?=,)',x)[0]

def anio(x):
    if pd.isnull(x):
        return np.nan
    else:
        return re.findall('\d*$',x)[0]
```

```python
nflx['mes']=nflx['Premiere'].apply(mes)
nflx['dia']=nflx['Premiere'].apply(dia)
nflx['anio']=nflx['Premiere'].apply(anio)
```

```python
nflx[nflx['Genre'].isna()]
```

#### Convertimos runtime a numero

```python
Actuals[Actuals['Genre'].isna()]
```

```python
x=Actuals['Runtime'][77]
y=Actuals['Runtime'][78]
print(x)
print(re.findall('\d*|\d*(?=–)',x))
print(y)
print(re.findall('\d*|\d*(?=–)',y))
```

```python
x=Actuals['Runtime'][77]
y=Actuals['Runtime'][78]
print(x)
print(re.search('–',x))
print(re.findall('\d*',x))
print(re.findall('(?<=–)\d*',x))
print(y)
print(re.search('–',y))
print(re.findall('\d*',y))
```

```python
def minimo_rt(x):
    if pd.isnull(x):
        return np.nan
    else:
        return re.findall('\d*',x)[0]
    
def maximo_rt(x):
    if pd.isnull(x):
        return np.nan
    else:
        if re.search('–',x):
            return re.findall('(?<=–)\d*',x)[0]
        else:
            return re.findall('\d*',x)[0]
```

```python
Actuals['min']=Actuals['Runtime'].apply(minimo_rt)
```

```python
Actuals['max']=Actuals['Runtime'].apply(maximo_rt)
```

#### Convertimos seasons a numeros

```python
x=Actuals['Seasons'][62]
y=Actuals['Seasons'][63]
print(x)
print(x.split(',')[1].strip().split())
print(y)
print(y.split(','))
```

```python
def temps(x):
    if pd.isnull(x):
        return np.nan
    else:
        return x.split(',')[0].split()[0]
    
def eps(x):
    if pd.isnull(x):
        return np.nan
    else:
        return x.split(',')[1].strip().split()[0]
```

```python
Actuals['Temps']=Actuals['Seasons'].apply(temps)
```

```python
Actuals['Eps']=Actuals['Seasons'].apply(eps)
```
